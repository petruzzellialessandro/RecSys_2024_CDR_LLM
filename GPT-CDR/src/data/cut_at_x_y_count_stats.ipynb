```python
import pandas as pd
import os
import sys
sys.path.append('src')
```

```python
def get_root_dir_path(curr_working_dir, no_folders_above):
    dir_path = curr_working_dir
    for i in range(no_folders_above):
        dir_path = os.path.dirname(dir_path)
    print(f"Initial directory: {curr_working_dir}\n"
          f"Root directory: {dir_path}."
          )
    return dir_path
```

```python
COLUMNS_TYPES =  {
        'user_id': 'string',
        'item_id': 'string',
        'rating': 'int',
        'timestamp': 'string',
        'title': 'string',
        'brand': 'string',
        'category': 'string'
    }
```

```python
ROOT_DIR = get_root_dir_path(os.getcwd(), 2)

DATA_DIR = ROOT_DIR + '/data/processed/'

BOOKS25_DIR = DATA_DIR + '/Books2_5/'
BOOKS35_DIR = DATA_DIR + '/Books3_5/'

CDS25_DIR = DATA_DIR + '/CDs2_5/'
CDS35_DIR = DATA_DIR + '/CDs3_5/'

MOVIES25_DIR = DATA_DIR + '/Movies2_5/'
MOVIES35_DIR = DATA_DIR + '/Movies3_5/'

DIR_LIST = [BOOKS25_DIR, BOOKS35_DIR,
            CDS25_DIR, CDS35_DIR,
            MOVIES25_DIR, MOVIES35_DIR]

EXTRA_DATA_DIR = ROOT_DIR + '/data/processed/extra_cut/'

EXTRA_BOOKS_DIR = EXTRA_DATA_DIR + '/Books_5_20/'
EXTRA_CDS_DIR = EXTRA_DATA_DIR + '/CDs_5_20/'
EXTRA_MOVIES_DIR = EXTRA_DATA_DIR + '/Movies_5_20/'
EXTRA_BOOKS_DIR1 = EXTRA_DATA_DIR + '/Books_5_30/'
EXTRA_CDS_DIR1 = EXTRA_DATA_DIR + '/CDs_5_30/'
EXTRA_MOVIES_DIR1 = EXTRA_DATA_DIR + '/Movies_5_30/'

EXTRA_DIR_LIST = [EXTRA_BOOKS_DIR, EXTRA_CDS_DIR, EXTRA_MOVIES_DIR, EXTRA_BOOKS_DIR1, EXTRA_CDS_DIR1, EXTRA_MOVIES_DIR1]
```

```python
def count_like_dislike(dataframe, dislike_up_to):
    
    if dislike_up_to == 3:
        print(f"Dislike ratings: 1, 2, 3 || Like ratings: 4, 5")
        rating_map = {1: 'Dislike', 2: 'Dislike', 3: 'Dislike', 4: 'Like', 5: 'Like'}
    elif dislike_up_to == 4:
        print(f"Dislike ratings: 1, 2, 3, 4 || Like ratings: 5")
        rating_map = {1: 'Dislike', 2: 'Dislike', 3: 'Dislike', 4: 'Dislike', 5: 'Like'}

    values = dataframe['rating'].value_counts()
    print(f"Before conversion to Like-Dislike:")
    for label, value in values.items():
      print(f"{label}: {value}")

    dataframe['rating'] = dataframe['rating'].map(rating_map)
    values = dataframe['rating'].value_counts()
    print(f"After conversion to Like-Dislike:")
    for label, value in values.items():
      print(f"{label}: {value}")
```

```python
def count_per_files_in_dir(dir_to_read, dislike_upper_lim):
    files = os.listdir(dir_to_read)
    for f in files:
        print(f"============= Dataset: {f} =================")
        df = pd.read_csv(dir_to_read + f, dtype=COLUMNS_TYPES)
        count_like_dislike(df, dislike_upper_lim)
        print(f"=========================================================")
```

```python
def drop_duplicates_df(directory_to_use):

    files = os.listdir(directory_to_use)

    for f in files:

        file_path = directory_to_use + f

        tmp_df = pd.read_csv(file_path)
        init_len = len(tmp_df)
        tmp_df.drop_duplicates(inplace=True)

        if init_len != len(tmp_df):
            print(f"============= File Path: {file_path} =================\n"
                  f"Found dups! New lenght: {len(tmp_df)}\nStoring dataframe.\n")
            tmp_df.to_csv(file_path)
            print(f"============= STORED DATAFRAME! =================\n"
                  f"============= Counting Likes&Dislike - UPPER LIM: 3! =================\n")
            count_like_dislike(tmp_df, 3)
            print(f"============= Counting Likes&Dislike - UPPER LIM: 4! =================\n")
            count_like_dislike(tmp_df, 4)
```

```python
for dir in EXTRA_DIR_LIST:
    print(f">>>>>>>>>>>>>>DIRECTORY: {dir}<<<<<<<<<<<<<")
    drop_duplicates_df(dir)
    print(f">>>>>>>>>>>>>> ------------------------------------- <<<<<<<<<<<<<\n")
```
